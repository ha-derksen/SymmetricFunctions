# Symmetric Functions

A standard reference for symmetric functions is:  
**[Mac]** I. G. MacDonald, *Symmetric Functions and Hall Polynomials*, 2nd edition, Oxford Mathematical Monographs.  

We will mostly stick the notation from this book.

We first load the module `Symmetric Functions` into `julia`. 

```julia
push!(LOAD_PATH, "../src")
using SymmetricFunctions
```

## Types of Symmetric Functions

The module introduces the new types `SymP`, `SymC` and `SymS` and the abstract type `Sym`.

### `SymP`

Symmetric functions are polynomials in infinitely many variables $x_1,x_2,x_3,\dots$ that are invariant under any permutation of the variables. The power sum $p_n=x_1^n+x_2^n+x_3^n+\dots$ is represented by the code `p_(n)`. If $\lambda=(\lambda_1,\lambda_2,\dots,\lambda_k)$ is a partition, then $p_\lambda$ is equal to the monomial $p_{\lambda_1}p_{\lambda_2}\cdots p_{\lambda_k}$ and is represented by the code `p_(λ)`. For example, we have `p_(3,3,2,1,1,1)=p_(3)*p_(3)*p_(2)*p_(1)*p_(1)*p_(1)`. The monomials $p_\lambda$ where $\lambda$ is a partition of $n$ form a basis of the space of homogeneous symmetric functions of degree $n$ (where $p_d$ has weighted degree $d$), which we will refer to as the $p$-basis or power sum basis. We will only work with homogeneous symmetric functions. Homogeneous symmetric functions expressed in the $p$-basis are of type `SymP`. We can construct symmetric functions of type `SymP` by taking linear combinations of $p$-monomials of the same degree:

```julia
f=3*p_(4,1)+5*p_(3,2)
```

```julia
typeof(f)
```

`f.degree` is the degree of the symmetric function `f`: 

```julia
f.degree
```

`f.coeffs` is a vector consisting of all coefficients of `f` in the $c$-basis. We always arrange the partitions of $n$ from small to large according to the pure lexicographic ordering. For example, for $n=5$ we have $(1,1,1,1,1)<(2,1,1,1)<(2,2,1)<(3,1,1)<(3,2)<(4,1)<(5)$.

```julia
f.coeffs
```

We can initialize a symmetric function in $p$-basis by using `SymP( )`, where the first argument is the degree and the second argument is a vector with the coefficients.

```julia
g=SymP(5,[1,2,3,4,5,6,7])
```

### `SymC`

For a partition $\lambda=(\lambda_1,\lambda_2,\dots,\lambda_k)$ one defines the scalar $z_\lambda$ as the size of the centralizer group of a permutation with cycle structure lambda in the symmetric group $S_n$, where $n=|\lambda|$. If we write $\lambda=(r^{m_r},(r-1)^{m_{r-1}},\dots,1^{m_1})$, where $m_i$ is the number of times that $i$ appears in $\lambda$, then we have $z_\lambda=\prod_{i\geq 1} (m_i!\cdot i^{m_i})$.
For example, $z_{(6,5,5,5,4,4)}=(1!6^1)(3!5^3)(2!4^2)=144000$.


```julia
z_(6,5,5,5,4,4)
```

 We define $c_\lambda=z_\lambda^{-1}p_\lambda$. The $c_\lambda$ where $\lambda$ is  a partition of $n$ form a basis of the symmetric functions of degree $n$ which we call the $c$-basis. Note that the $c$-basis is just a rescaling of the $p$-basis. The $c$-basis does not appear explicitly in [Mac], but it is useful because sometimes using the $c$-basis instead of the $p$-basis avoids the use of rational numbers instead of integers. Symmetric functions expressed in the $c$-basis are of type `SymC`. We use the letter $c$ because the $c$-basis elements correspond to conjugacy classes.

### `SymS`

For a partition $\lambda$ we denote the corresponding Schur symmetric function by $s_\lambda$. The $s_\lambda$ where $\lambda$ is a partition of $n$ form the $s$-basis or Schur basis of homogeneous symmetric functions of degree $n$.

### `Sym`

The book [Mac] discusses several other bases, such as the $h$-basis, $e$-basis (using elementary symmetric functions) etc. These are not (yet?) implemented. The abstract type `Sym` is a homogeneous symmetric function in any basis, i.e., it is the union of the types `SymP`, `SymC` and `SymS`.

## Conversions

The conversion between `SymS`  and `SymP`/`SymC` lies at the heart of the algorithm. These conversions can be computed quickly by a clever recursive use of the Murnaghan-Nakayama rule (and even more quickly by using multi-threading if there is more than one core).

### Conversions with `SymP( )`, `SymC( )` and `SymS( )`

We can easily convert between symmetric functions of type `SymP`, `SymC` and `SymS` using the functions `SymP( )`, `SymC( )` and `SymS( )`. For example:

```julia
f=3*c_(5)+2*c_(4,1)-3*c_(2,2,1)
g=SymP(f)
```

```julia
SymC(g)
```

### Characters values of the symmetric group

If we express a schur function $s_\lambda$ in the $c$-basis whose elements correspond to conjugacy classes, then we get the character values of the irreducible character $\chi^\lambda$ of $S_n$. 

```julia
f=SymC(s_(3,1,1))
```

```julia
f.coeffs
```

### Dimensions of the irreducible characters of the symmetric group

If we convert $p_{(1,1,1,\dots,1)}=p_{(1^n)}$ into the Schur basis, then the coefficients are the dimensions of the irreducible representations (which can also be computed by the hook length formula). For example, we get the dimensions of the irreducible representations of $S_6$ by:

```julia

SymS(p_(1,1,1,1,1,1))
```

## Multiplication of symmetric functions

If `f` and `g` are homogeneous symmetric functions in the same basis, then `f*g` is their product (expressed in the same basis).

### Multiplication in the $p$-basis

In the $p$-basis, the multiplication is just multiplication of polynomials in the variables $p_1,p_2,p_3,\dots$. For example, $(p_{(2)}+p_{(1,1)})(p_{(3)}+p_{(2,1)})=(p_2+p_1^2)(p_3+3p_2p_1)=6p_2p_1^3+3p_2^2p_1+2p_3p_1^2+p_3p_2=6p_{(2,1,1,1)}+3p_{(2,2,1)}+2p_{(3,1,1)}+p_{(3,2)}$:

```julia
(p_(2)+2*p_(1,1))*(p_(3)+3*p_(2,1))
```

### Tensor products of representations of the general linear group

Suppose that $V$ is an $t$-dimensional ${\mathbb C}$-vector space. For every partition  $\lambda=(\lambda_1,\lambda_2,\dots,\lambda_k)$ with $k\leq t$
we have an irreducible polynomial representation $S_\lambda(V)$ of  the group  ${\rm GL}(V)\cong {\rm GL}_t({\mathbb C})$.
(If $k>t$ then we set $S_\lambda(V)=0$.) A homogeneous symmetric function $f=\sum_{\lambda\vdash n} \alpha_\lambda s_\lambda$ in the $s$-basis of degree $n$ with nonnegative integer coefficients corresponds to the representation $F=\bigoplus_{\lambda \vdash n} S_\lambda(V)^{\alpha_\lambda}$.
Similarly,  $g=\sum_{\mu\vdash m} \beta_\mu s_\mu$ corresponds to the representation $G=\bigoplus_{\lambda \vdash m} S_\lambda(V)^{\beta_\lambda}$.
The product $fg$ expressed in the Schur basis, corrsponds to the tensor product $F\otimes G$. If $fg=\sum_{\nu\vdash n+m} \gamma_\nu s_\mu$, then 
$F\otimes G=\bigoplus_{\nu\vdash n+m} S_\nu(V)^{\gamma_\mu}$ is the decomposition into irreducible ${\rm GL}(V)$-representations.
In the special case where $f=s_\lambda$ and $g=s_\mu$ are Schur basis elements with $\lambda\vdash n$ and $\mu\vdash m$, then $s_\lambda s_\mu=\sum_{\nu\vdash n+m}c_{\lambda,\mu}^\nu s_{\nu}$ and $S_\lambda(V)\otimes S_\mu(V)=\bigoplus_{\nu\vdash n+m}S_\nu(V)^{c_{\lambda,\mu}^\nu}$, where $c_{\lambda,\mu}^\nu$ is the Littlewood-Richardson coefficient. For example, to decompose $S_{(3,1)}(V)\otimes S_{(4,1,1)}(V)$ we compute:


```julia
s_(3,1)*s_(4,1,1)
```

## Tensor products of representations of the symmetric group

Let $\chi^\lambda$, $\lambda\vdash n$ be the irreducible characters of the symmetric group $S_n$. These characters are functions
from the set $\{K_\mu\mid \mu\vdash n\}$ of conjugacy classes of $S_n$ to ${\mathbb Z}$.
If $f=\sum_{\lambda\vdash n}\alpha_\lambda s_\lambda$ is a nonnegative integer combination, then $f$ corresponds to a character $\sum_{\lambda \vdash n} \alpha_\lambda\chi^\lambda$ of the symmetric group and is again a function from the set of conjugacy classes of $S_n$ to ${\mathbb Z}$.
If we write $f$ in the $c$-basis, say $f=\sum_{\mu\vdash n}\beta_\mu c_\mu$, then $\beta_\mu=f(K_\mu)$.
Now $c_\lambda$ is the class function that is equal to $1$ on the conjugacy class $K_\lambda$ and equal to $0$ on all conjugacy classes $K_\mu$ with $\mu\neq \lambda$.


### The operation `&` in the $c$-basis


We introduce a binary operation `&` of symmetric functions of degree $n$ that corresponds to the product of class functions of $S_n$. The elements $c_\lambda$, $\lambda\vdash n$ are orthogonal idempotents. In particular, we have $(\sum_{\lambda\vdash n}\alpha_\lambda c_\lambda)(\sum_{\lambda\vdash n}\beta_\lambda c_\lambda)=\sum_{\lambda\vdash n}(\alpha_\lambda\beta_\lambda)c_\lambda$.

```julia
(2*c_(1,1,1)+3*c_(2,1)+5*c_(3))&(7*c_(1,1,1)+11*c_(2,1)+13*c_(3))
```

### The operation `&` in the $s$-basis

In the $s$-basis, the operation `&` computes tensor product decompositions of products of representations of $S_n$. If $\lambda,\mu$ are partitions of $n$, then $\chi^\lambda\chi^\mu$ decomposes as $\sum_{\mu\vdash n} g_{\lambda,\mu}^\nu \chi^\mu$, where $g_{\lambda,\mu}^\nu$ is a Kronecker coefficient. For example, to decompose $\chi^{(3,2,1)}\chi^{(4,1,1)}$ as a sum of irreducible characters we compute:

```julia
s_(3,2,1)&s_(4,1,1)
```
